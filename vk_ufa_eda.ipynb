# Определяем фичи и создаем данные для мужчин и женщин

```python
fields=['country', 'city','bdate','career','education', 'has_photo', 'sex', 'movies','verified','can_see_all_posts','can_send_friend_request','quotes']
```

```python
data_woman = vk.users.search(count = 1000, country=1, city=151, fields=fields, sex=1)['items']
data_man = vk.users.search(count = 1000, country=1, city=151, fields=fields, sex=2)['items']
```

```python
len(data_woman)
```

```python
data_woman[0].keys()
```

```python
data = []
```

```python
for i in range(len(data_woman)):
    if (("city" in data_woman[i].keys()) and ("bdate" in data_woman[i].keys() and data_woman[i]['bdate'].count('.')==2)):
        if (data_woman[i]['city']['title'] == 'Уфа'):
            data.append(data_woman[i])
```

```python
for i in range(len(data_man)):
    if ("city" in data_man[i].keys()) and ("bdate" in data_man[i].keys() and data_man[i]['bdate'].count('.')==2):
        if (data_man[i]['city']['title'] == 'Уфа'):
            data.append(data_man[i])
```

```python
len(data)
```

# Создание датасета

```python
import pandas as pd

df=pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['career'])
df.head()
```

```python
df['city'] = 'Уфа'
df['country'] = 'Россия'
```
